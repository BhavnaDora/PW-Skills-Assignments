##### Min-Max scaling or Normalization is a feature scaling technique used for data pre processing. It transforms data by scaling features in the range (0,1). The formula (xi-xmin)/(xmax-xmin) is applied to every data point and the new data point obtained lies in the range (0,1).

```python
# example of min-max scaling-
import seaborn as sns
import pandas as pd
```

```python
df= sns.load_dataset("tips")
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max= MinMaxScaler()        # min_max is an object of MinMaxScaler class and is initialized

pd.DataFrame(min_max.fit_transform(df[["total_bill"]]), columns=["toatl_bill_scaled"])
```

##### Unit vector technique is a type of feature scaling method in which a row of data is viewed as a D- dimensional vector, and it is scaled (shrinked/stretched) to unit length. When used on the entire dataset, the transformed data can be visualized as a bunch of vectors with different directions on the D-dimensional unit sphere. Unit vector is obtained as= vector/magnitude of that vector

#### (3)-
##### The Principal Component Analysis is a popular unsupervised learning technique for reducing the dimensionality of data. It increases interpretability yet, at the same time, it minimizes information loss. It helps to find the most significant features in a dataset and makes the data easy for plotting in 2D and 3D.
##### PCA can be used for a variety of purposes, including data visualization, feature selection, and data compression. In data visualization, PCA can be used to plot high-dimensional data in two or three dimensions, making it easier to interpret. In feature selection, PCA can be used to identify the most important variables in a dataset. In data compression, PCA can be used to reduce the size of a dataset without losing important information.
##### PCA helps in dimensionality reduction by finding a line/plane called principal component that best fits the data points and then projects the data points onto that line/plane. So, in this way it reduces the dimension while capturing most of the variance of the data. PCA employs a linear transformation that is based on preserving the most variance in the data using the least number of dimensions. 

```python
# example-
import pandas as pd
from sklearn.datasets import load_breast_cancer

#instantiating-
cancer= load_breast_cancer()

#creating dataframe-
df1= pd.DataFrame(cancer["data"], columns= cancer["feature_names"])
df1
```

```python
df1.shape
```

```python
# feature scaling-
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()         # scaler is an object of StandardScaler class and is initialized
scaled_data= scaler.fit_transform(df1)
```

```python
# dimensionality reduction using PCA-
from sklearn.decomposition import PCA

pca= PCA(n_components=2)    #initializing PCA object
x_pca= pca.fit_transform(scaled_data)

x_pca.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer["target"])
plt.xlabel("first principal component")
plt.ylabel("second principal component")
```

#### (4)-
##### PCA is unsupervised learning algorithm that is used for feature extraction and feature selection, where it selects the most important features in the dataset and makes the data easy to visualize in 2D and 3D. PCA performs the following steps for feature extraction-
- Standardize the dataset
- Construct the covariance matrix
- Perform Eigendecomposition of covariance matrix
- Selection of most important Eigenvectors / Eigenvalues
- Projection matrix creation of important eigenvectors
- Training / test dataset transformation
##### Using sklearn.decomposition library's PCA class, we can directly perform the dimensionality reduction and selection of important features, we just have to give the no. of components we want as argument.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins_data= sns.load_dataset("penguins")
penguins_data.head()
```

```python
penguins_data.info()
```

```python
penguins_data.isnull().sum()
```

```python
for col in penguins_data.select_dtypes(include="float", exclude="object"):
    penguins_data[col].fillna(penguins_data[col].mean(), inplace=True)
    
```

```python
penguins_data["sex"].fillna(method="ffill", inplace= True)
```

```python
penguins_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in penguins_data[['island', 'sex']]:
    penguins_data[col] = LabelEncoder().fit_transform(penguins_data[col])
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

for col in penguins_data[["bill_length_mm",	"bill_depth_mm", "flipper_length_mm", "body_mass_g"]]:
    penguins_data[col]= scaler.fit_transform(penguins_data[[col]])
```

```python
from sklearn.decomposition import PCA
x= penguins_data.drop("species", axis=1)
y= penguins_data["species"]
pca= PCA()
x_pca= pd.DataFrame(pca.fit_transform(x))
x_pca
```

```python
pd.DataFrame(pca.explained_variance_ratio_).plot.bar()
plt.xlabel('Principal Components')
plt.ylabel('Explained Varience')
```

```python
# the first 3 PC capture most variance of the data, so the 6 original features can be replaced with these 3 components.
# so the 6 dimensions have been reduced to 3 dimensions.
```

```python
pca.components_
```

```python
pd.DataFrame(pca.components_, columns= x.columns, index=["PC1", "PC2", "PC3", "PC4", "PC5", "PC6"])
```

##### PCA is unsupervised learning alfgorithm used for dimensionality reduction and selection of important features. It increases interpretability yet, at the same time, it minimizes information loss. It helps to find the most significant features in a dataset and makes the data easy for plotting in 2D and 3D. 
##### PCA helps in dimensionality reduction by finding a line/plane called principal component that best fits the data points and then projects the data points onto that line/plane. So, in this way it reduces the dimension while capturing most of the variance of the data. PCA employs a linear transformation that is based on preserving the most variance in the data using the least number of dimensions. 
##### Dimensionality reduction of dataset using PCA-
- First, we do pre processing od the dataset and standardize it. 
- Then we split the dataset into train and test set.
- construct the covariance matrix of train set
- Then we perform eigendecomposition of the covariance matrix into eigen vectors and eigen values. Eigenvectors represent the principal components and eigenvalues represent their magnitude.
- Using explained variance ratio we select the most important features i.e. the top k eigenvectors having large eigenvalues are selected. So, we have k dimensional feature subsapce.
- construct projection matrix- The k eigenvectors selected are used for constructing the projection matrix.
- Transform the training data i.e. the d dimensional input is tranformed to k dimensional feature subspace. Hence, dimensionality has been reduced using PCA.
