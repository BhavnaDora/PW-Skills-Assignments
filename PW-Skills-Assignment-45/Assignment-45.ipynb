```python
import pandas as pd
import numpy as np

np.random.seed(123)
tv_hours= np.random.uniform(2,6,50)
tv_hours
```

```python
physical_activity= ["medium", "high", "high", "medium", "medium", "medium", "low", "low", "low", "medium", 
                    "medium", "low", "medium", "high", "high", "medium", "high", "high", "low", "medium",
                    "medium", "low", "low", "medium", "low", "high", "high", "high", "medium", "low", 
                    "high", "medium", "medium", "low", "medium", "high", "medium", "low", "high", "low",
                    "low", "high", "high", "high", "low", "medium", "low", "low", "high", "high"]

df3= pd.DataFrame({"tv_hours": tv_hours, "physical_activity": physical_activity})
df3
```

```python
# encoding categorical variable physical_activity using OrdinalEncoder-
from sklearn.preprocessing import OrdinalEncoder
ord_encoder= OrdinalEncoder(categories=[["low","medium","high"]])
df3["phy_activity_encoded"]= ord_encoder.fit_transform(df3[["physical_activity"]])
df4= df3.drop("physical_activity", axis=1)
```

```python
df4
```

```python
df4.corr(method="pearson")
```

```python
# so, the pearson correlation coefficient= -0.647, which is high negative correlation, 
# shows that lesser the tv_hours higher is the level of physical activity.
```

#### (5)-

```python
df5= pd.DataFrame({"age": [25, 42, 37, 19, 31, 28], "soft_drink": ["coke", "pepsi", "mountain dew", "coke", "pepsi", "coke"]})
df5
```

```python
# encoding the categorical variable soft_drink using OneHotEncoder-
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder()
encoded= encoder.fit_transform(df5[["soft_drink"]]).toarray()
encoded
```

```python
encoded_df= pd.DataFrame(encoded, columns= encoder.get_feature_names_out())
df6= pd.concat([df5.drop("soft_drink", axis=1), encoded_df], axis=1)
df6
```

```python
df6.corr()
```
