```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
lst= np.linspace(1,10,100)
```

```python
mean= np.mean(lst)
mean
```

```python
std= np.std(lst)
std
```

```python
def normal_distribution(lst, mean, std):
    prob_density=[]
    for i in lst:
        y= (1/np.sqrt(2.0*np.pi*(std**2.0)))* (np.exp(-(i-mean)**2/(2.0*std**2)))
        prob_density.append(y)
    
    return prob_density
```

```python
pdf= normal_distribution(lst, mean, std)
```

```python
plt.plot(lst,pdf)
plt.xlabel("data points")
plt.ylabel("probability density")
```

#### (4)-
##### Binomial distribution with parameters n and p is the discrete probability distribution of the number of successes in a sequence of n independent experiments, each asking a yes-no question and each with its own boolean valued outcome: success(with probability p) and failure(with probability 1-p).
##### The no. of trials n is fixed
##### The probability of success and failure remains constant for each trial.
##### For a single trial binomial distribution becomes bernoulli distribution.
##### Examples-
##### Tossing a fair coin 3 times and finding probability of getting 2 heads
##### Rolling a fair dice 10 times and finding probability of getting 3

```python
x= range(1,10)
```

```python
mean= np.mean(x)
mean
```

```python
def cumulative_distribution(x,mean):
    cdf=0.0
    for i in x:
        cdf= cdf+ (np.exp(-mean)*(mean**i)/np.math.factorial(i))
        
    return cdf
```

```python
cumulative_distribution(x,mean)
```

#### (7)-
##### Binomial distribution is the discrete probability distribution having binary outcomes for a process repeated n no. of times. Poisson distribution is the discrete probability distribution of no. of events occuring in a fixed time interval.
##### Binomial distribution is biparametric (n and p), whereas poisson distribution is uniparametric (mean).
##### There are fixed no. of trials in binomial distribution but unlimited no. of trials in poisson distribution.
##### Binomial distribution has binary outcomes- success and failure but Poisson distribution can have unlimited no. of possible outcomes.
##### in binomial distribution mean> variance but in poisson distribution mean=variance.
##### Poisson distribution is the limit of binomial distribution for large n. When the no. of events is high but the probability of success is low, poisson distribution is applied.

#### (9)-
##### In binomial distribution, Mean= np and variance= npq, where n is the no. of trials, p is the probability of success and q is the probability of failure. so,  mean > variance
##### In poisson distribution, mean= lambda * t , variance= lambda * t where lambda is the expected no. of events occuring in every time interval t. so, mean = variance
