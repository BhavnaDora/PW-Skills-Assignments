##### Estimation statistics is a data analysis framework that uses a combination of effect sizes, confidence intervals, precision planning, and meta-analysis to plan experiments, analyze data and interpret results. Estimate is a specified observed numerical value used to estimate an unknown population parameter.
##### Point estimate- It is a single numerical value used to estimate an unknown population parameter. Ex) sample mean is a point estimate of population mean.
##### Interval estimate- It is a range of values used to estimate an unknown population parameter.

#### (2)-

```python
import numpy as np
def estimate_population_mean(sample_mean, sample_std, sample_size):
    z= 1.96  # considering a 95% confidence interval
    margin_of_error= z* sample_std/np.sqrt(sample_size)
    lower_CI= sample_mean- margin_of_error
    upper_CI= sample_mean+ margin_of_error
    return margin_of_error, lower_CI, upper_CI
```

```python
# the population mean is 95% likely to fall within the interval bounded by lower-bound and upper-bound

MOE, lower_bound, upper_bound= estimate_population_mean(20, 2.5, 35)
print("the margin of error is", MOE)
print("the lower bound of confidence interval is", lower_bound)
print("the upper bound of confidence interval is", upper_bound)
```

#### (3)-
##### Hypothesis testing is a type of statistical analysis in which we put our assumptions about a population parameter to the test. It is used to estimate the relationship between 2 statistical variables. 
##### Using the sample data we draw conclusions/inferences about some unknown population parameter like mean, variance, standard deviation etc. of the population data. To do this we use hypothesis testing.
##### Hypothesis testing is important because we can draw inferences about population parameters which we do by taking sample data rather than population data because entire population data is impossible to obtain. Hypothesis testing helps the researcher to successfully extrapolate data from the sample to the larger population. It allows us to evaluate the strength of our claim or assumption before implementing it in our data set.

#### (4)-
##### Null hypothesis (H0)- It is the beginning assumption that the average weight of male students = average weight of female students. mean weight(males) = mean weight(females)
##### Aletrnate hypothesis(H1)- It is opposite to null hypothesis. We have to see whether average weight of male students is greater than that of female students. So, mean weight(males) > mean weight(females)
##### Experiment/Statistical tests- Here we can use 2 sample T-test. We define significance level and also calculate the degree of freedom, using them in t table we calculate critical value. We then compare the t value obtained from 2 sample T-test stats with critical value.
##### Conclusion- If t value is greater than critical value we reject the null hypothesis, if t value is less than critical value we fail to reject the null hypotheis.

##### Null hypothesis (H0)- It is the beginning assumption that the 2 population means are equal. mean1 = mean2
##### Aletrnate hypothesis(H1)- It is opposite to null hypothesis. mean1 ≠ mean2

#### (6)-
##### Null hypothesis- Null hypothesis suggests that there is no relationship between the two variables. Generally, researchers try to reject or disprove it. Represented as H0. Null hypothesis is rejected if the p-value is less than the alpha-value; otherwise, it is accepted.
##### Alternate hypothesis- Alternative hypothesis is a method for reaching a conclusion and making inferences and judgements about certain facts or a statement. It is opposite to Null hypothesis. Represented as H1. An alternative hypothesis is accepted if the p-value is less than the alpha-value otherwise, it is rejected.
##### Example- The average height of students is 168 cm. But on measuring it was different from 168cm. So, here- Null hypothesis H0- Mean = 168 cm. Alternate hypothesis H1- Mean != 168 cm

#### (7)- 
##### Steps in Hypothesis testing-
##### Step 1- Specify the Null Hypothesis
##### Step 2- Specify the Alternate Hypothesis
##### Step 3- Set the Significance Level alpha- It is usually 0.05 and confidence interval is 95%
##### Step 4- Calculate the Test Statistic and Corresponding P-Value
##### Step 5- Drawing a Conclusion- P-value <= significance level => Reject your null hypothesis. P-value > significance level => Fail to reject your null hypothesis.

#### (8)-
##### P-value is a number calculated from a statistical test, that describes how likely you are to have found a particular set of observations if the Null hypothesis were true. The p-value describes the probability of obtaining a sample statistic as or more extreme by chance alone if your null hypothesis is true. P values are used in hypothesis testing to help decide whether to reject the Null hypothesis. The smaller the p-value, the greater the evidence against the null hypothesis. Ex) P-value = 0.01 This will happen 1 in 100 times by pure chance if your null hypothesis is true.

```python
# 2- sample T-test-
import numpy as np

# Null hypothesis H0- Population means are equal i.e. mean1 = mean2
# Alternate hypothesis H1- mean1 != mean2

sample1= np.random.normal(10, 1, 100)
sample2= np.random.normal(9, 0.5, 100)

# These 2 samples follow normal distribution as we took random samples (of same sample size) from a normal distribution
# before applying 2 sample t-test we have to check if the variances are similar or not i.e. the ratio must be less than 4:1 
# then we can say the 2 variances are similar and we will apply 2 sample t test for equal variances
```

```python
var1= np.var(sample1)
var2= np.var(sample2)
var1/var2
```

```python
from scipy.stats import ttest_ind

significance_value= 0.05      # considering a 95% confidence interval
t_value, p_value= ttest_ind(sample1, sample2, equal_var= True)
print("t statistics value is", t_value)
print("p value is", p_value)

# checking whether to accept or reject Null hypothesis-
if p_value < significance_value:
    print("Reject the Null hypothesis")
else:
    print("Fail to reject the Null hypothesis")
```

```python
import numpy as np

sample1_mean= 80
sample1_std= 10
n1= 30              # this is sample 1 size

sample2_mean= 75
sample2_std= 8
n2= 40              # this is sample 2 size

significance_level= 0.01
```

##### Null hypothesis(H0)- The populatuion means of two groups are equal. mean1 = mean2
##### Alternate hypothesis(H1)- The population means of two groups are not equal i.e. mean1 != mean2
##### We will apply 2 sample t test to check whether the population means of 2 groups are equal or not. 

```python
sample1_var= sample1_std**2
sample2_var= sample2_std**2
sample1_var/sample2_var
```

##### so we can consider the sample variances to be similar as the ratio is less than 4:1. So we apply 2 sample t test with equal variances.

```python
# calculating difference of means- 
diff_of_means= sample1_mean - sample2_mean

# calculating pooled standard deviation- 
pooled_var= ((n1-1)*sample1_var + (n2-1)*sample2_var)/(n1+n2-2)
pooled_std= np.sqrt(pooled_var)

# calculating t test stats-
t= diff_of_means/ (pooled_std* np.sqrt((1/n1) + (1/n2)))
print("the t value of 2 sample t test is", t)
```

```python
tot_df= (n1-1)+(n2-1)
tot_df
```

##### using the total degree of freedom and significance level for 2 tail test in t table we get -
##### t_table_value= +-(2.6501)

```python
t_table_value= 2.6501
```

```python
if t < t_table_value:
    print("Fail to reject Null hypothesis")
else:
    print("Reject Null hypothesis")
```

##### Therefore as t lies in the region from -2.6501 to +2.6501, we fail to reject the Null hypothesis. This means the population measn of the 2 groups are equal.
